<h3> Задача1</h3>
<p> Найти собственные векторы и собственные значения для линейного оператора, заданного матрицей

$$A=\begin{pmatrix}
-1 & -6\\ 
2 & 6
\end{pmatrix}.$$
</p>

```
# Найдем собственные значения из характеристического уравнения
# -1 - L  -6
#  2   6-L  = 0
# 
#(-1-L)*(6-l)-(-6*2) = 0
#-6-6L+L+L^2+12 = 0
#L^2-5L+6 = 0
# По теореме Виетта сумма корней - 5 (-1*второй коэффициент), произведение корней свободный член - 6
# L1 = 2 , L2 = 3
```

```
# Найдем собственные векторы для оператора
# L1 = 2
#  -1-2 -6 = 0
#   2   6-2 = 0
#
#-3x1 - 6x2 =0
#2x1 + 4x2 = 0
# x1 = -2x2 
#x2 = -1
#x1 = 2

# L2 = 3 
# -1 -3  6 =0
# 2      6-3
# -4x1 + 6x2 =0 
# 2x1 + 3x2 = 0 
# 2x1 = -3x2 
# x1 = -1.5x2
# x2 = -2
 # x1 = 3
# Ответ собственные векторы (2, -1) (3; -2)
# Собственные числа 2 и 3 
```

```
# Проверка
A = np.array([[-1,-6], [2,6]])

w,v = np.linalg.eig(A)
print('Собственные числа ', w)
print('Собственные вектора', v)
```

<h3> Задача2</h3>
<p> Дан оператор поворота на 180 градусов, задаваемый матрицей 

$$A=\begin{pmatrix}
-1 & 0\\ 
0 & -1
\end{pmatrix}.$$

Показать, что __любой__ вектор является для него собственным.
</p>

```
# Найдем собственные числа оператора
w2, v2 = np.linalg.eig(np.array([[-1,0], [0, -1]]))
print('Собственные числа оператора ', w2)
```

```
# Собвтенное чило оператор равно -1, составим уравнение для векторов
# -1 - 1  0 =0
# 0   -1-1  = 0
#
# 0x1 + 0x2 =0
# 0x1 + 0x2 =0 
# Таким образом для любых значений x1,x2 являются решением этой системы уравлений, 
# соответсвенно любой вектор для данного оператора будет собственным.
```

<h3> Задача3</h3>
<p> Пусть линейный оператор задан матрицей

$$A=\begin{pmatrix}
1 & 1\\ 
-1 & 3
\end{pmatrix}.$$

Установить, является ли вектор $x=(1,1)$ собственным вектором этого линейного оператора.
</p>

```
# Если вектор собственный, тогда должно существовать число l, при котором A * x = lx
# проверим
# (1 *1) + (1 *1 )  = l*1
# (-1 *1) + (3 * 1 ) = l*1
# 2 = l
# 2 = l - система уравлений имеет смысл, l = 2 существует, вектор x является собственным
```
